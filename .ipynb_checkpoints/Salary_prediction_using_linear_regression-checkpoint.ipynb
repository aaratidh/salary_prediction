```python
import numpy as np 
import matplotlib.pyplot as ptl 
import pandas as pd 
```

```python
data = pd.read_csv('salary.csv')
```

```python
data
```

```python
data.dtypes
```

```python
from sklearn.model_selection import train_test_split 
from pattern.en import number 
```

```python
#replacing Nan with the O value
data =data.replace(np.nan,0)
data["experience"]
```

```python
#convert non-numerical variable into numercal value 
exp_colm = data["experience"]
for i in range  (0,len(exp_colm)):

    if type(exp_colm[i])!= int:
        exp_colm[i] = number(exp_colm[i])
    
    
        
   
```

```python
#adding updated value in datframe
data['experience']=exp_colm
```

```python
#now data is ready for machine learning
data
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

```python
Y_test_pred = model.predict(X_test)
Y_test_pred
```
